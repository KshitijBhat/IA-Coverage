```python
path = []
for i in range(800,99,-1):
    path.append((200,i))
    
```

```python
class Robot:
    radius = 30
    bot_color = 'r'
    
    def __init__(self,x,y,yaw):
        self.x = x
        self.y = y
        self.yaw = yaw
    
    def plot(self,ax):
        bot_circle = plt.Circle( (self.x, self.y),self.radius,color=self.bot_color)
        arrow = ax.arrow(self.x,self.y,self.radius*np.cos(self.yaw),self.radius*np.sin(self.yaw))
        #patch = ax.add_patch(bot_circle)
        return [bot_circle,arrow]
```

```python
robot = Robot(200,800,4.7124)
fig,ax = plt.subplots(figsize=(8,8))
ax.imshow(arena,cmap = plt.cm.Dark2,origin = 'lower')
robot.plot(ax)
for xi,yi in path:
    ax.scatter(xi, yi, marker='.', color='r',s=1)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
%matplotlib inline

plt.rcParams["animation.html"] = "jshtml"
from IPython.display import HTML


fig, ax = plt.subplots()


def f(x, y):
    return np.sin(x) + np.cos(y)

x = np.linspace(0, 2 * np.pi, 120)
y = np.linspace(0, 2 * np.pi, 100).reshape(-1, 1)

# ims is a list of lists, each row is a list of artists to draw in the
# current frame; here we are just animating one artist, the image, in
# each frame
ims = []
for i in range(60):
    x += np.pi / 15.
    y += np.pi / 20.
    im = ax.imshow(f(x, y), animated=True)
    if i == 0:
        ax.imshow(f(x, y))  # show an initial one first
    ims.append([im])

ani = animation.ArtistAnimation(fig, ims, interval=50, blit=True,
                                repeat_delay=1000)
plt.close()
ani
```
