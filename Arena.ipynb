```python
un_image = cv2.imread('un_arena.png')
```

```python
img_bw = cv2.cvtColor(un_image,cv2.COLOR_BGR2GRAY)
```

```python
imgrid = np.zeros((1024,1024))
for i in range(1024):
    for j in range(1024):
        if img_bw[i,j] >= 230:
            imgrid[i,j] = 1
        if img_bw[i,j] <230:
            imgrid[i,j] = 0
            
```

```python
fig, ax = plt.subplots(figsize=(12,12))
ax.imshow(imgrid, cmap=plt.cm.gray)
ax.scatter(200,800, marker = ".", color = "yellow", s = 200)
ax.scatter(100,150, marker = ".", color = "yellow", s = 200)
ax.scatter(660,200, marker = ".", color = "yellow", s = 200)
ax.scatter(800,600, marker = ".", color = "yellow", s = 200)
ax.scatter(400,1000, marker = ".", color = "yellow", s = 200)
#ax.scatter(goal[1],goal[0], marker = "+", color = "red", s = 200)
```

```python
A = (200,800)
B = (200,100)
C = (660,200)
D = (800,600)
E = (400,1000)
AB_x = [A[0],B[0]]
AB_y = [A[1],B[1]]
BC_x = [B[0],C[0]]
BC_y = [B[1],C[1]]
CD_x = [C[0],D[0]]
CD_y = [C[1],D[1]]
DE_x = [D[0],E[0]]
DE_y = [D[1],E[1]]
EA_x = [E[0],A[0]]
EA_y = [E[1],A[1]]
```

```python
fig, ax = plt.subplots(figsize=(12,12))
ax.imshow(imgrid, cmap=plt.cm.gray)
ax.scatter(200,800, marker = ".", color = "green", s = 200,label = "Start")
# ax.scatter(200,100, marker = ".", color = "yellow", s = 200)
# ax.scatter(660,200, marker = ".", color = "yellow", s = 200)
ax.scatter(400,1000, marker = ".", color = "yellow", s = 200,label = "Stop")
ax.plot(AB_x,AB_y,'r',label = "Reference Path")
ax.plot(BC_x,BC_y,'r')
ax.plot(CD_x,CD_y,'r')
ax.plot(DE_x,DE_y,'r')
ax.legend()
#ax.plot(EA_x,EA_y)
plt.savefig("Arena.png")
```
